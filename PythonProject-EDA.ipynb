#### Utilizando groupby e aggregate functions:
Seguimos um processo chamado “split-apply-combine” para agrupar os dados e analisar as características agregadas ou combinadas de cada grupo:



1.   Divida os dados em grupos com base em alguns critérios (neste exemplo, dividimos os dados com base nos critérios de status do cliente);
2.   Aplique a função para agregar e analisar cada grupo de dados. Você pode especificar os dados que deseja analisar (neste exemplo, utilizamos apenas a contagem de clientes inativos e ignoramos o restante dos dados);
3.   Combine os resultados em uma tabela de dados ou dataframe.



```python
# primeiro iremos agrupar a coluna ('ifood_status' == 'Inactive') utilizando .groupby('ifood_status').get_group('Inactive')
inactive_customers = df_customer_segmentation.groupby('ifood_status').get_group('Inactive')

# depois iremos agrupar por .groupby('segmentation_month') e realizar a contagem
# dos clientes inativos, e apresentar em ordem decrescente os meses com os 
# maiores números de inativos
(
    inactive_customers
      .groupby('segmentation_month')
      .agg(top3_meses_clientes_inativos = ('ifood_status', 'count'))
      .nlargest(3, 'top3_meses_clientes_inativos')
)
```

*Como podemos analisar, os 3 meses com maior número de clientes inativos se concentram no ano de 2019.*

*Os meses são:* 
  - Outubro; 

  - Novembro; 

  - Dezembro. 

**Quais são os 5 pratos mais preferidos do cliente do tipo Marlin?**

Será necessário criar uma função chamada `recuperar_preferred_dishes` para recuperar corretamente os dados. Essa função deve retornar uma lista com os pratos preferidos.

A variável 'preferred_dishes' apresenta os campos em formato de array. Esses campos precisam ser transformados.

ATENÇÃO - Considere um único tipo de prato por cliente, por exemplo, o cliente com id 1a2b3c4d5e tem 6 vezes comida brasileira, deverá ser considerado como apenas 1.

```python
# primeiro iremos realizar filtros nas colunas que queremos ('customer_id', 'marlin_tag', 'preferred_dishes')
# depois iremos executar query do campo '1. Marlin', referente a variável 'marlin_tag'
marlin_preferred_dishes = df_customer_segmentation.filter(['customer_id', 'marlin_tag', 'preferred_dishes']).query("marlin_tag == '1. Marlin'")
marlin_preferred_dishes
```

```python
# criando a função para tratamento da variável 'preferred_dishes'
# utilizaremos o .replace() e o .split() para limpar o array
def recuperar_preferred_dishes(df):
  df = str(df)
  df1 = df.replace("\\", "")
  df2 = df1.replace('"', "")
  df3 = df2.replace("[", "")
  df4 = df3.replace("]", "")
  df5 = df4.split(",")
  
  return df5
```

```python
# aplicando a função de tratamento para a variável 'preferred_dishes'
marlin_preferred_dishes['preferred_dishes'] = marlin_preferred_dishes['preferred_dishes'].apply(recuperar_preferred_dishes)
marlin_preferred_dishes
```

```python
# utilizando a função explode para quebrar o array em novas linhas
# para o mesmo 'customer_id'
marlin_preferred_dishes_explode = marlin_preferred_dishes.explode('preferred_dishes')
marlin_preferred_dishes_explode
```

```python
# considerando um único tipo de prato por cliente, 
# utilizaremos o método .nunique() para retornar únicos 
# dentro do mesmo 'customer_id' 
marlin_preferred_dishes_explode.nunique()

# agora para realizar o cálculo dos 5 pratos mais pedidos do 
# cliente do tipo '1. Marlin':
# primeiro iremos agrupar utilizando .groupby('preferred_dishes')
# depois iremos utilizar .agg() para criar uma coluna e passar 
# 'preferred_dishes' e 'count' como parâmetros
# por fim, exibiremos o 5 pratos mais pedidos utilizando .nlargest()
(
    marlin_preferred_dishes_explode
      .groupby('preferred_dishes')
      .agg(top5_most_preferred_dishes = ('preferred_dishes', 'count'))
      .nlargest(5, 'top5_most_preferred_dishes')
)
```

**Realizando o passo a passo abaixo:**

* 1. Faça concatenação das variáveis `user_identifier` e `session_id`, nesta ordem, para criar um novo identidicador único, chamado `unique_id`

* 2. Insira a coluna `unique_id` na posição 0 do DataFrame `df_session_visits`. Utilize uma variável auxiliar para facilitar.

* 3. Crie uma nova coluna com a diferença entre as variáveis `session_started_at_utc0` e `session_ended_at_utc0`. O nome da coluna deve ser `tempo_sessao`.

* 4. Crie uma função que irá receber dois parâmetros, uma DataFrame e outra a quantidade de usuário mais ativos dentro da plataforma.

```python
# passo 1: concatenação das variáveis 'user_identifier' e 'session_id'
df_sessions_visits['unique_id'] = df_sessions_visits['user_identifier'] + df_sessions_visits['session_id']
df_sessions_visits['unique_id']
```

```python
# passo 2: set o index da coluna 'unique_id' para 0
df_sessions_visits['unique_id'] = df_sessions_visits['user_identifier'] + df_sessions_visits['session_id']
df_sessions_visits.set_index('unique_id', inplace = True)
df_sessions_visits.head(2)
```

```python
# convertendo as colunas em pd.datetime
df_sessions_visits['session_started_at_utc0'] = pd.to_datetime(df_sessions_visits['session_started_at_utc0'])
df_sessions_visits['session_ended_at_utc0'] = pd.to_datetime(df_sessions_visits['session_ended_at_utc0'])

# passo 3: nova coluna com a diferenca entre 'session_started_at_utc0' e 'session_ended_at_utc0'
df_sessions_visits['tempo_sessao'] = df_sessions_visits['session_ended_at_utc0'] - df_sessions_visits['session_started_at_utc0']
df_sessions_visits['tempo_sessao'] = df_sessions_visits['tempo_sessao'] / np.timedelta64(1,'m') 
df_sessions_visits['tempo_sessao']
```

```python
# passo 4: Crie uma função que irá receber dois parâmetros, uma DataFrame e outra a quantidade de usuário mais ativos dentro da plataforma.
# função simples que contém parâmetro do DataFrame e outro que armazenará o número especificado pelo usuário
def most_active_customers(df, n):
  df = df.sort_values(by=['tempo_sessao'], ascending=False).head(n)

  return df
```
